# "The only free lunch in finance"
Say we have security A and B:
* A has a volatility of 10 and a return of 4 
* B has a volatility of 14 and a return of 6

Intuitively, the return of a 50-50 portfolio is the weighted average, i.e. 5.

However, the volatilty depends on the correlation between A and B.
* If they are perfectly correlated (1), the volatility would be the weighted average
* The combination if they are not correlated is less than expected (in this case 10.1). 

_There is a nose to the curve of the weighting whereby the volaitlity of the portfolio is less than the volatility of either one_
