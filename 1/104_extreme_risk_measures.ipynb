## The Gaussian Assumption
* The standard simplifying assumotion is that asset returns are normally distributed
* In reality, large changes are more frequent

## Higher order moments
In order to do the actual returns justice, we need to look at skewness and kurtoic distributions


N.B: Kurtosis > 3 == fat-tail

* Most asset returns are non-normal
* The deviations from normaility can be very severe
* Using volatility as a risk estimate for a negatively skew distribution may lead to underestimate the tail risk associated to strong negative returns
