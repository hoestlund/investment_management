```python
# NFLX occurs 101 times, let's see if it mentioned in any other tweets.
# Lets investigate what other stocks were mentioned when NFLX was the symbol being discussed

for tweet in tweets[tweets['symbols']=='NFLX']['text']:
    print(re.findall('\$[a-zA-Z]+', tweet))
```

```python
# Applying this to the dataframe
tweets_df = tweets[tweets['symbols']=='NFLX']['text'].apply(lambda tweet: re.findall('\$[a-zA-Z]+', tweet))
tweets_df
```

```python
# Then convert it to a boolean index signifying connections by mapping arrays to series
matrix = tweets_df.apply(lambda contained: pd.Series(1, index=set(map(str.upper, contained)))).fillna(0)
matrix
```

```python
# Most commonly cross referenced for Netflix
cross_reference = matrix.sum()
cross_reference
```
